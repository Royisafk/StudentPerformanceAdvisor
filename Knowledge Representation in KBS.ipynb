```python
#Example of Knowledge Representation in KBS
class Person:
    def _init_(self, name, age):
        self.name = name
        self.age = age

    def _str_(self):
        return f"{self.name}, {self.age} years old"

class KnowledgeBase:
    def _init_(self):
        self.people = []

    def add_person(self, person):
        self.people.append(person)

    def find_person_by_name(self, name):
        for person in self.people:
            if person.name == name:
                return person
        return None
# Example usage
if _name_ == "_main_":    kb = KnowledgeBase()
kb.add_person(Person("Alice", 30))

person = kb.find_person_by_name("Alice")
if person:
        print(person)
else:
        print("Person not found.")
```
