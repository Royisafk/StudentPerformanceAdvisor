```python
{
  "rules": {
    "Scholarship": {
      "conditions": [
        "GPA > 3.5",
        "Attendance > 80",
        "No Disciplinary Cases"
      ],
      "conclusion": "Eligible for Scholarship"
    },

    "Graduation": {
      "conditions": [
        "GPA > 3.0",
        "Completed Prerequisites",
        "No Outstanding Fees"
      ],
      "conclusion": "Eligible for Graduation"
    },

    "Probation": {
      "conditions": [
        "GPA < 3.0"
      ],
      "conclusion": "Academic Probation"
    },

    "Registration": {
      "conditions": [
        "Outstanding Fees"
      ],
      "conclusion": "Registration Blocked"
    },

    "DeansList": {
      "conditions": [
        "GPA > 3.5",
        "Attendance > 80"
      ],
      "conclusion": "Dean's List Candidate"
    }
  }
}

class AcademicAdvisor:
    def __init__(self):
        self.conclusions = []
        self.explanations = {}
    def infer(self, student):
        self.conclusions = []
        self.explanations = {}
        if (student["gpa"] > 3.5 and
            student["attendance"] > 80 and
            not student["disciplinary_cases"]):
            self.conclusions.append("Eligible for Scholarship")
            self.explanations["Eligible for Scholarship"] = [
                "GPA > 3.5",
                "Attendance > 80%",
                "No Disciplinary Cases"
            ]
        if (student["gpa"] > 3.0 and
            student["completed_prerequisites"] and
            not student["outstanding_fees"]):
            self.conclusions.append("Eligible for Graduation")
            self.explanations["Eligible for Graduation"] = [
                "GPA > 3.0",
                "Completed Prerequisites",
                "No Outstanding Fees"
            ]
        if student["gpa"] < 3.0:
            self.conclusions.append("Academic Probation")
            self.explanations["Academic Probation"] = [
                "GPA < 3.0"
            ]
        if student["outstanding_fees"]:
            self.conclusions.append("Registration Blocked")
            self.explanations["Registration Blocked"] = [
                "Outstanding Fees"
            ]
        if (student["gpa"] > 3.5 and
            student["attendance"] > 80):
            self.conclusions.append("Dean's List Candidate")
            self.explanations["Dean's List Candidate"] = [
                "GPA > 3.5",
                "Attendance > 80%"
            ]
        return self.conclusions
    def explain(self):
        print("\nExplanation Facility")
        for conclusion, reasons in self.explanations.items():
            print(f"\n{conclusion} because:")
            for reason in reasons:
                print(f"- {reason}")
            print("Therefore the rule was activated.")
gpa = float(input("Enter GPA: "))
attendance = float(input("Enter Attendance Percentage: "))
disciplinary = input("Any disciplinary cases? (yes/no): ").strip().lower()
prerequisites = input("Completed prerequisites? (yes/no): ").strip().lower()
fees = input("Outstanding fees? (yes/no): ").strip().lower()
student = {
    "gpa": gpa,
    "attendance": attendance,
    "disciplinary_cases": disciplinary == "yes",
    "completed_prerequisites": prerequisites == "yes",
    "outstanding_fees": fees == "yes"
}
advisor = AcademicAdvisor()
results = advisor.infer(student)
print("\nConclusions:")
if results:
    for result in results:
        print("✓", result)
else:
    print("No conclusions generated.")
advisor.explain()
```
