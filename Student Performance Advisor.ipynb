```python
class KnowledgeBase:

    def __init__(self):
        self.facts = {}

    def add_fact(self, fact):
        self.facts[fact] = True

    def has_fact(self, fact):
        return self.facts.get(fact, False)


kb = KnowledgeBase()
print("STUDENT PERFORMANCE ADVISOR")

# Fact 1 & 2
attendance = input("Attendance good? (yes/no): ")
if attendance == "yes":
    kb.add_fact("attendance_good")
else:
    kb.add_fact("attendance_poor")

# Fact 3 & 4
assignment = input("Assignments completed? (yes/no): ")
if assignment == "yes":
    kb.add_fact("assignment_complete")
else:
    kb.add_fact("assignment_incomplete")

# Fact 5 & 6
study = input("Study hours high? (yes/no): ")
if study == "yes":
    kb.add_fact("study_hours_high")
else:
    kb.add_fact("study_hours_low")

# Fact 7 & 8
exam = input("Exam score high? (yes/no): ")
if exam == "yes":
    kb.add_fact("exam_high")
else:
    kb.add_fact("exam_low")

# Fact 9 & 10
motivation = input("Motivation high? (yes/no): ")
if motivation == "yes":
    kb.add_fact("motivation_high")
else:
    kb.add_fact("motivation_low")

# Fact 11 & 12
participation = input("Class participation high? (yes/no): ")
if participation == "yes":
    kb.add_fact("participation_high")
else:
    kb.add_fact("participation_low")

# Fact 13 & 14
project = input("Project completed? (yes/no): ")
if project == "yes":
    kb.add_fact("project_completed")
else:
    kb.add_fact("project_not_completed")

# Fact 15
time_management = input("Good time management? (yes/no): ")
if time_management == "yes":
    kb.add_fact("time_management_good")

print("\nRULE EVALUATION")

# Rule 1
if kb.has_fact("attendance_good") and kb.has_fact("assignment_complete"):
    kb.add_fact("academic_engagement")
    print("Rule 1 Fired: Academic Engagement")

# Rule 2
if kb.has_fact("study_hours_high") and kb.has_fact("motivation_high"):
    kb.add_fact("good_preparation")
    print("Rule 2 Fired: Good Preparation")

# Rule 3
if kb.has_fact("attendance_good") and kb.has_fact("study_hours_high"):
    kb.add_fact("consistent_learner")
    print("Rule 3 Fired: Consistent Learner")

# Rule 4
if kb.has_fact("assignment_complete") and kb.has_fact("study_hours_high"):
    kb.add_fact("prepared_student")
    print("Rule 4 Fired: Prepared Student")

# Rule 5
if kb.has_fact("attendance_good") and kb.has_fact("motivation_high"):
    kb.add_fact("active_student")
    print("Rule 5 Fired: Active Student")

# Rule 6
if kb.has_fact("attendance_poor") and kb.has_fact("assignment_incomplete"):
    kb.add_fact("poor_engagement")
    print("Rule 6 Fired: Poor Academic Engagement")

# Rule 7
if kb.has_fact("study_hours_low") and kb.has_fact("exam_low"):
    kb.add_fact("weak_preparation")
    print("Rule 7 Fired: Weak Preparation")

# Rule 8
if kb.has_fact("motivation_low") and kb.has_fact("study_hours_low"):
    kb.add_fact("low_commitment")
    print("Rule 8 Fired: Low Commitment")

# Rule 9
if kb.has_fact("project_completed"):
    kb.add_fact("practical_skills")
    print("Rule 9 Fired: Practical Skills Developed")

# Rule 10
if kb.has_fact("participation_high"):
    kb.add_fact("classroom_engagement")
    print("Rule 10 Fired: Classroom Engagement")

# Rule 11
if kb.has_fact("time_management_good") and kb.has_fact("study_hours_high"):
    kb.add_fact("effective_study_habits")
    print("Rule 11 Fired: Effective Study Habits")

# Rule 12 (Multi-step)
if kb.has_fact("academic_engagement") and kb.has_fact("good_preparation"):
    kb.add_fact("high_success_potential")
    print("Rule 12 Fired: High Success Potential")

# Rule 13 (Multi-step)
if kb.has_fact("high_success_potential") and kb.has_fact("exam_high"):
    kb.add_fact("excellent_performance")
    print("Rule 13 Fired: Excellent Academic Performance")

# Rule 14
if kb.has_fact("poor_engagement") and kb.has_fact("weak_preparation"):
    kb.add_fact("risk_of_failure")
    print("Rule 14 Fired: Risk of Failure")

# Rule 15
if kb.has_fact("classroom_engagement") and kb.has_fact("practical_skills"):
    kb.add_fact("well_rounded_student")
    print("Rule 15 Fired: Well-Rounded Student")

print("\nCONCLUSION")

if kb.has_fact("excellent_performance"):
    print("Conclusion: Excellent Academic Performance")
    print("Recommendation: Continue your excellent study habits and pursue advanced academic goals.")

elif kb.has_fact("risk_of_failure"):
    print("Conclusion: Risk of Failure")
    print("Recommendation: Improve attendance, study hours, and assignment completion.")

elif kb.has_fact("well_rounded_student"):
    print("Conclusion: Well-Rounded Student")
    print("Recommendation: Maintain balanced academic and practical learning.")

elif kb.has_fact("exam_high"):
    print("Conclusion: Strong Academic Performance")
    print("Recommendation: Maintain current study habits.")

else:
    print("Conclusion: Average Academic Performance")
    print("Recommendation: Increase study hours and participation.")
```
